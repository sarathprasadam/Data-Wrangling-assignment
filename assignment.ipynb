## Filter the dataframe with only this columns
   ['continent','location','date','total_cases','total_deaths','gdp_per_capita','
   human_development_index'] and update the data frame.


```python
df=df.loc[:, ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df.head()
```

# Data Cleaning

## Remove all duplicates observations

```python
df.duplicated().sum()
```

SINCE NO DUPLICATES NO NEED TO REMOVE

## Find missing values in all columns

```python
df.isnull().sum()
```

 ## Remove all observations where continent column value is missing

```python
df_cont=df.dropna(subset=["continent"])
df_cont
```

```python
df_cont.isnull().sum()
```

## Fill all missing values with 0

```python
df_no_null=df_cont.fillna(0)
df_no_null
```

```python
df_no_null.isnull().sum()
```

# CaseStudy-part2

# Date time format 

## Convert date column in datetime format using pandas.to_datetime

```python
df1=df_no_null
df1.head()
```

```python
df1['date']=pd.to_datetime(df1['date'])
df1.head(2)
```

## Create new column month after extracting month data from date

```python
df1['month']=df1['date'].dt.month
df1.head(3)
```

# Data Aggregation

## Find max value in all columns using groupby function on 'continent'
       


```python
df2=df1.groupby('continent').max()
df2
```

```python
df2.index
```

## Store the result in a new dataframe named 'df_groupby'

```python
df_groupby=df2
df_groupby
```

# Feature Engineering 

## Create a new feature 'total_deaths_to_total_cases' by ratio of
        'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```

```python
df_groupby=df_groupby.reset_index()
```

```python
df_groupby
```
